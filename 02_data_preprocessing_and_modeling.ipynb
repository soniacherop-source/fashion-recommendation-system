# Data Preprocessing

In this section, we prepare the dataset for collaborative filtering by transforming and organizing the data into a suitable format for recommendation modeling. This includes checking data consistency, encoding categorical variables, and creating the user-item interaction matrix.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('data/fashion_products.csv')
```

## Encoding Categorical Features

Machine learning algorithms work better with numerical data. Therefore, categorical variables such as Product Name, Brand, Category, Color, and Size are converted into numerical representations using label encoding.

```python
label_encoder = LabelEncoder()

df['Product Name Encoded'] = label_encoder.fit_transform(df['Product Name'])

df['Brand Encoded'] = label_encoder.fit_transform(df['Brand'])

df['Category Encoded'] = label_encoder.fit_transform(df['Category'])

df['Color Encoded'] = label_encoder.fit_transform(df['Color'])

df['Size Encoded'] = label_encoder.fit_transform(df['Size'])

df.head()
```

### Interpretation

The categorical features were successfully transformed into numerical values using label encoding. This preprocessing step makes the dataset suitable for machine learning algorithms and recommendation system modeling.

## Creating User-Item Interaction Matrix

Collaborative filtering relies on user-product interactions. Therefore, a user-item matrix is created where rows represent users, columns represent products, and values represent product ratings.

```python
user_item_matrix = df.pivot_table(
    index='User ID',
    columns='Product ID',
    values='Rating'
)

user_item_matrix.head()
```

```python
user_item_matrix.isnull().sum().sum()
```

## Handling Missing Values

The user-item matrix contains many missing values because users have not rated every product. These missing values are replaced with zeros to prepare the matrix for similarity calculations.

```python
user_item_matrix_filled = user_item_matrix.fillna(0)

user_item_matrix_filled.head()
```

### Interpretation

Missing ratings were replaced with zeros to indicate the absence of interaction between users and products. This allows similarity calculations to be performed efficiently during collaborative filtering.

## Computing User Similarity

Cosine similarity is used to measure similarity between users based on their product ratings. Users with similar rating patterns are likely to have similar fashion preferences.

```python
user_similarity = cosine_similarity(user_item_matrix_filled)

user_similarity.shape
```

The similarity matrix contains 100 users compared against one another, resulting in a 100 × 100 matrix. Each value represents the similarity score between two users based on their product rating behavior.

```python
user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix_filled.index,
    columns=user_item_matrix_filled.index
)

user_similarity_df.head()
```

### Interpretation

The similarity matrix shows how closely users are related based on their rating behavior.Higher similarity scores indicate users with similar product preferences and rating patterns. These similarities are later used to generate personalized product recommendations.

The dataset was successfully prepared for collaborative filtering by encoding categorical variables, constructing the user-item interaction matrix, handling missing values, and computing user similarity scores. These preprocessing steps ensure that the data is structured appropriately for recommendation model development.

# Generating Product Recommendations

In this section, we build a user-based collaborative filtering recommendation system. The system identifies users with similar preferences and recommends products based on the ratings and interactions of similar users.

```python
def recommend_products(user_id, num_recommendations=5):

    similar_users = user_similarity_df[user_id].sort_values(ascending=False)

    similar_users = similar_users.drop(user_id)

    most_similar_user = similar_users.index[0]

    products_rated_by_similar_user = user_item_matrix_filled.loc[most_similar_user]

    products_not_rated_by_user = user_item_matrix_filled.loc[user_id] == 0

    recommended_products = products_rated_by_similar_user[products_not_rated_by_user]

    recommended_products = recommended_products.sort_values(ascending=False)

    return recommended_products.head(num_recommendations)
```

```python
recommend_products(1)
```

## Interpretation

The recommendation function successfully generates personalized product recommendations for a selected user by identifying similar users and suggesting products they highly rated. This demonstrates how collaborative filtering can provide customized fashion recommendations based on shared user preferences.

## Displaying Recommended Product Details

```python
def get_recommended_product_details(user_id, num_recommendations=5):

    recommended_product_ids = recommend_products(user_id, num_recommendations).index

    recommended_products = df[df['Product ID'].isin(recommended_product_ids)][
        ['Product ID', 'Product Name', 'Brand', 'Category', 'Price', 'Rating']
    ]

    return recommended_products
```

```python
get_recommended_product_details(1)
```

### Interpretation

The recommendation system successfully returns detailed fashion products for a selected user based on collaborative filtering. Instead of displaying only product IDs, the system now provides meaningful product information such as product name, brand, category, price, and rating. This improves interpretability and makes the recommendation results more useful and user-friendly.

## Evaluating Recommendation Results

Collaborative filtering systems can be evaluated by analyzing recommendation outputs and similarity patterns between users. In this section, we visualize user similarities and recommendation rating distributions.

```python
plt.figure(figsize=(10,8))

sns.heatmap(user_similarity_df.iloc[:10, :10], annot=True)

plt.title('User Similarity Heatmap')

plt.show()
```

### Interpretation

The heatmap visualizes similarity scores between users based on their product ratings. Darker or higher values indicate stronger similarity between users. This confirms that the collaborative filtering model can identify users with related fashion preferences, which improves recommendation quality.

```python
recommended_products = get_recommended_product_details(1)

plt.figure(figsize=(8,5))

plt.bar(
    recommended_products['Product Name'],
    recommended_products['Rating']
)

plt.title('Ratings of Recommended Products')

plt.xlabel('Product Name')

plt.ylabel('Rating')

plt.show()
```

### Interpretation

The chart shows the ratings of products recommended to the selected user. Most recommended products have relatively high ratings, indicating that the recommendation system prioritizes highly-rated fashion items from similar users.
