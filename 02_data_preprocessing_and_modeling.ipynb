# Data Preprocessing

In this section, we prepare the dataset for collaborative filtering by transforming and organizing the data into a suitable format for recommendation modeling. This includes checking data consistency, encoding categorical variables, and creating the user-item interaction matrix.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('data/fashion_products.csv')
```

## Encoding Categorical Features

Machine learning algorithms work better with numerical data. Therefore, categorical variables such as Product Name, Brand, Category, Color, and Size are converted into numerical representations using label encoding.

```python
label_encoder = LabelEncoder()

df['Product Name Encoded'] = label_encoder.fit_transform(df['Product Name'])

df['Brand Encoded'] = label_encoder.fit_transform(df['Brand'])

df['Category Encoded'] = label_encoder.fit_transform(df['Category'])

df['Color Encoded'] = label_encoder.fit_transform(df['Color'])

df['Size Encoded'] = label_encoder.fit_transform(df['Size'])

df.head()
```

### Interpretation

The categorical features were successfully transformed into numerical values using label encoding. This preprocessing step makes the dataset suitable for machine learning algorithms and recommendation system modeling.

## Creating User-Item Interaction Matrix

Collaborative filtering relies on user-product interactions. Therefore, a user-item matrix is created where rows represent users, columns represent products, and values represent product ratings.

```python
user_item_matrix = df.pivot_table(
    index='User ID',
    columns='Product ID',
    values='Rating'
)

user_item_matrix.head()
```

```python
user_item_matrix.isnull().sum().sum()
```

## Handling Missing Values

The user-item matrix contains many missing values because users have not rated every product. These missing values are replaced with zeros to prepare the matrix for similarity calculations.

```python
user_item_matrix_filled = user_item_matrix.fillna(0)

user_item_matrix_filled.head()
```

### Interpretation

Missing ratings were replaced with zeros to indicate the absence of interaction between users and products. This allows similarity calculations to be performed efficiently during collaborative filtering.

## Computing User Similarity

Cosine similarity is used to measure similarity between users based on their product ratings. Users with similar rating patterns are likely to have similar fashion preferences.

```python
user_similarity = cosine_similarity(user_item_matrix_filled)

user_similarity.shape
```

The similarity matrix contains 100 users compared against one another, resulting in a 100 × 100 matrix. Each value represents the similarity score between two users based on their product rating behavior.

```python
user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix_filled.index,
    columns=user_item_matrix_filled.index
)

user_similarity_df.head()
```

### Interpretation

The similarity matrix shows how closely users are related based on their rating behavior.Higher similarity scores indicate users with similar product preferences and rating patterns. These similarities are later used to generate personalized product recommendations.

The dataset was successfully prepared for collaborative filtering by encoding categorical variables, constructing the user-item interaction matrix, handling missing values, and computing user similarity scores. These preprocessing steps ensure that the data is structured appropriately for recommendation model development.
